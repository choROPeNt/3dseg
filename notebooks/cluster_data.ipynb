```python
import numpy as np
import nrrd
import os
import matplotlib.pyplot as plt
import skimage 
```

```python
file = "data/200_10-layer_00_0_1_24_04_29.seg.nrrd"
```

```python
seg , header_seg = nrrd.read(file)
```

```python
plt.imshow(seg[0,::])
```

```python
array = seg[::] == 2
```

```python
plt.imshow(array[0,::])
```

```python
labels = skimage.measure.label(array,connectivity=1)
print(np.unique(labels))
```

```python
plt.imshow(labels[0,::] ==2)
```

```python
for i in range(len(np.unique(labels))):
    print(np.sum(labels==i ))
```

```python
shape = labels.shape

# Define the NRRD header
header = {
    'type': 'int',  # Data type of the segmentation data
    'dimension': 3,  # Number of dimensions
    'sizes': shape,  # Size of each dimension
    'space': 'left-posterior-superior',  # Coordinate space
    # Add any other necessary metadata here
}


# Write the data to the NRRD SEG file
nrrd.write("test2.seg.nrrd", labels, header)
```
