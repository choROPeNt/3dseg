```python
import numpy as np

def extract_subvolumes(data, subvolume_shape, stride):
    z_stride, x_stride, y_stride = stride
    z_size, x_size, y_size = subvolume_shape

    subvolumes = []
    locations = []

    for z in range(0, data.shape[0] - z_size + 1, z_stride):
        for x in range(0, data.shape[1] - x_size + 1, x_stride):
            for y in range(0, data.shape[2] - y_size + 1, y_stride):
                subvolume = data[z:z + z_size, x:x + x_size, y:y + y_size]
                subvolumes.append(subvolume)
                locations.append((z, x, y))

    return np.array(subvolumes), np.array(locations)

# Example usage
data = np.random.random((256, 256, 256))  # Replace this with your 3D data
subvolume_shape = [64, 64, 64]  # Replace this with the desired subvolume shape
stride = [16, 16, 16]  # Replace this with the desired stride

subvolumes, locations = extract_subvolumes(data, subvolume_shape, stride)

for subvolume, location in zip(subvolumes, locations):
    print(f"Subvolume at location {location}:\n{subvolume.shape}")
```

```python
from torch3dseg.datasets.utils import SliceBuilder
```

```python
builder = SliceBuilder(
    raw_dataset=data,
    label_dataset=None,
    weight_dataset=None,
    patch_shape=subvolume_shape,
    stride_shape=stride)
```

```python
slices = builder.raw_slices
```
