```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = "plotting/val_loss_avg"
fig, axs = plt.subplots(figsize=(4.8,4.8))

for file in os.listdir(path):
    file_name = os.path.splitext(file)[0]

    df = pd.read_csv(os.path.join(path,file))

    axs.plot(df["Step"],df["Value"],label=file_name)

axs.legend()
axs.grid()
axs.set_xlabel("iterations")
axs.set_ylabel("val_loss_avg")
# axs.set_yscale('log')
fig.tight_layout()
plt.savefig("plotting/val_loss_avg.pdf",dpi=300)
```
